```
# ==========================================================================
# Libraries

# ======== Libraries ===================
from __future__ import absolute_import, division, print_function, unicode_literals

try:
  # %tensorflow_version only exists in Colab.
  %tensorflow_version 2.x
except Exception:
  pass



import tensorflow as tf
print( tf.constant( 'TensorFlow version :  ' + tf.__version__ ) )
print(' ============================================== ')

import numpy as np
from numpy import array
import pandas as pd

import tensorflow as tf
from tensorflow import keras 

from tensorflow.keras import layers
from keras.layers import LSTM
from keras.layers import RepeatVector
from keras.layers import Dropout
from sklearn import preprocessing

import matplotlib.pyplot as plt
import datetime as dt
import os
import glob
import sys
import shutil

# ==== confirm  that we can connect to the GPU with tensorflow ====
# First, you'll need to enable GPUs for the notebook: Navigate to Edit→Notebook Settings
device_name = tf.test.gpu_device_name()
#if device_name != '/device:GPU:0':
#  raise SystemError('GPU device not found')
print('GPU connection confirmed with: {}'.format(device_name))
print(' =========== ')
# ==== confirm  that we can connect to the GPU with tensorflow ====

# ==== change the default size of the Matplotlib graph ===
#fig_size = plt.rcParams["figure.figsize"]
#fig_size[0] = 15
#fig_size[1] = 15
#plt.rcParams["figure.figsize"] = fig_size
# ==== change the default size of the Matplotlib graph ===

# ==========================================================================
# split a multivariate sequence into samples
def split_sequences(sequences, n_steps_in, n_steps_out):
	X, y = list(), list()
	for i in range(len(sequences)):
		# find the end of this pattern
		end_ix = i + n_steps_in
		out_end_ix = end_ix + n_steps_out
		# check if we are beyond the dataset
		if out_end_ix > len(sequences):
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequences[i:end_ix, :], sequences[end_ix:out_end_ix, :]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)

# ==========================================================================
def get_col(arr, col):
    return map(lambda x : x[col], arr)

# ==========================================================================
def extractYfinance(symboltoextract, mykeyfigure):
    import yfinance as yf
    # import pandas as pd
    #symboltoextract = '^VIX'
    #mykeyfigure = "Close"
    my_api_extract = yf.Ticker(symboltoextract)
    # fetch data by interval (including intraday if period < 60 days)
    # valid intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo
    # (optional, default is '1d')
    mydatagranularity = '1d'
    myvalues1= my_api_extract.history(period="5y", interval = mydatagranularity) #my_api_extract.history(period="5y")
    myvalues2 = myvalues1.rename(index=str, columns={mykeyfigure: symboltoextract + '_' + mykeyfigure})
    #myInfo1=my_api_extract.info, ['Close']
    Myextract1= myvalues2.loc[:, ([symboltoextract + '_' + mykeyfigure])]
    return Myextract1    
# ==========================================================================
# ==========================================================================


def Predictmyshare(myshare):

    MyinvestigatexIndex = myshare  # GT CON.DE GOOG AMZN VRSN CL IP SAN.PA T ABCA.PA MSFT
    Myextract1 = extractYfinance(MyinvestigatexIndex, "High") # that is the element we will predict
    Myextract2 = extractYfinance(MyinvestigatexIndex, "Close")
    Myextract3 = extractYfinance(MyinvestigatexIndex, "Open")
    Myextract4 = extractYfinance(MyinvestigatexIndex, "Low")
    Myextract5 = extractYfinance(MyinvestigatexIndex, "Volume")
    Myextract6 = extractYfinance("^VIX", "Close")     # ^VIX   ^EXIV
    Myextract7 = extractYfinance("SPY", "Close")      # SPY    ^GDAXI
    Myextract8 = extractYfinance("SPY", "Volume")     # SPY    ^GDAXI
    Originaldata= pd.concat([Myextract1,Myextract2,Myextract3,Myextract4,Myextract5,Myextract6,Myextract7,Myextract8], axis=1, sort=False)
    # === kill line with empty cells
    Originaldata.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
    now=dt.datetime.now()
    mytimestamp = now.strftime("%Y%m%d%H%M") 
    Originaldatafilename = mytimestamp + "_" + MyinvestigatexIndex.replace(".","")  + "_history"  + ".csv"
    # Originaldata.to_csv('/content/drive/My Drive/FromColab/' + Originaldatafilename, index=True)
    # ==========================================================================

    # ==========================================================================
    # define input
    Mynameofcolumntopredict= Originaldata.columns.values[0]
    Mydataset = Originaldata
    NumberOfVariables = len(Mydataset.columns)-1   # xx number of columns (after the date) to be forecasted
    Mydataset = (Mydataset.iloc[:, 1:(NumberOfVariables+1)].values)  # 5 columns, date is ignored and used as index
    NumberofHistoricalrecords = Mydataset.shape[0] # number of records in the dataset; shape[1]  # gives number of col count

    myscaler = preprocessing.MinMaxScaler()
    Mydataset = myscaler.fit_transform(Mydataset)

    # ==========================================================================
    # covert into input/output
    X, y = split_sequences(Mydataset, n_steps_in, n_steps_out)
    n_features = X.shape[2] # the dataset knows the number of features, e.g. 2


    # /////////// Testing /////////
    train_size = int(len(Mydataset) * Percentageofdatasetfortraining / 100)
    test_size = len(Mydataset) - train_size
    train = Mydataset[0:train_size,:]
    test = Mydataset[train_size:len(Mydataset),:]
    print("train size: {}, test size: {} ".format(len(train), len(test)))

    trainX, trainY = split_sequences(train, n_steps_in, n_steps_out)  
    testX, testY = split_sequences(test, n_steps_in, n_steps_out)

    print('shape of trainX: ', trainX.shape)
    print('shape of testX: ', testX.shape)


    # ///////////


    # ==========================================================================
    # define model
    model = []
    model = tf.keras.Sequential()
    model.add(keras.layers.LSTM(NumberofNeurons, activation='relu', input_shape=(n_steps_in, n_features)))
    model.add(keras.layers.RepeatVector(n_steps_out))
    model.add(keras.layers.LSTM(NumberofNeurons, activation='relu', return_sequences=True))

    # Adding a second and third layer and some Dropout regularisation
    model.add(keras.layers.LSTM(int(NumberofNeurons/4), activation='relu', return_sequences=True))
    model.add(keras.layers.LSTM(int(NumberofNeurons/8), activation='relu', return_sequences=True))

    model.add(keras.layers.TimeDistributed(keras.layers.Dense(n_features)))

    model.compile(optimizer='adam', loss='mse', metrics=['acc'])


    # Print a summary of the model's architecture
    model.summary()

    # fit model
    # history = model.fit(X, y, epochs=TrainingSets, verbose=1 )

    import time
    start_time=time.time()
    history = model.fit(trainX, trainY, batch_size=mybatchsize, epochs=TrainingSets, verbose=1, validation_data=(testX, testY))
    #history = model.fit(trainX, trainY, epochs=TrainingSets, verbose=1, validation_data=(testX, testY))
    print('Training took {} seconds'.format(time.time()-start_time))

    myacc= history.history['val_acc'][-1:]
    myacc1=myacc[0]
    myacc2=int(myacc1*10000)/100
    print('accuracy')
    print(myacc2)

    # === Plot accuracy ================
    plt.figure(1)  
   
    # summarize history for accuracy  
    plt.subplot(211)  
    plt.plot(history.history['acc'])  
    plt.plot(history.history['val_acc'])  
    plt.title('model accuracy')  
    plt.ylabel('accuracy')  
    plt.xlabel('epoch')  
    plt.legend(['train', 'test'], loc='upper left')  
   
     # summarize history for loss  
    plt.subplot(212)  
    plt.plot(history.history['loss'])  
    plt.plot(history.history['val_loss'])  
    plt.title('model loss')  
    plt.ylabel('loss')  
    plt.xlabel('epoch')  
    plt.legend(['train', 'test'], loc='upper left')  
    plt.show() 
    # === Plot accuracy ================



    # demonstrate prediction
    Mydataset_input = (pd.DataFrame(Mydataset).iloc[:, 0:(NumberOfVariables)].tail(n_steps_in))   # xx last columns to be used to generate estimates
    Mydataset_input = np.array(Mydataset_input)
    x_input = np.array(Mydataset_input)
    x_input1 = x_input.reshape((1, n_steps_in, n_features))
    yhat = model.predict(x_input1, verbose=1)
    # print('=== np.array(Mydataset_input) ===')
    # print(np.array(Mydataset_input))
    # print('==================')
    # print('=== yhat ===')
    # print(yhat[:,:,:])
    # print('==================')
    predicted_Values =yhat[:,:,0]  # [:3, 1:2]  # this is an array
    PV= np.array(predicted_Values)
    prediction=np.transpose(PV)
    print (prediction)

    print('=== scaled back values ===')
    MyActuals = Originaldata
    MyActuals = (MyActuals.iloc[:, 1:2].values)  # Plot the first  variable
    x_max = np.amax(MyActuals)
    x_min= np.amin(MyActuals)
    scaledbackprediction = pd.DataFrame.from_records(prediction)
    scaledbackprediction = scaledbackprediction * (x_max - x_min) + x_min
    scaledbackprediction = scaledbackprediction.round(decimals= 2) # round 2 digits 
    print(' //////////// ')
    print(Mynameofcolumntopredict)
    print(' //////////// ')
    print(scaledbackprediction)            # should be the predictions scaled back



    #=== make a summary file ===   
    Mysharename= pd.DataFrame([Mynameofcolumntopredict],columns=['Share'])
    now=dt.datetime.now()
    Mytime = now.strftime("%Y%m%d__%H_%M_%S") 
    Mytimepd= pd.DataFrame([Mytime],columns=['Timestamps'])
    dataforanalysis = pd.concat([Mysharename, Mytimepd, scaledbackprediction])

    # if file does not exist write header 
    if not os.path.exists('/content/drive/My Drive/FromColab/' + "'AA___Summary_TS2v1.csv'"):
        dataforanalysis.to_csv('/content/drive/My Drive/FromColab/' + 'AA___Summary_TS2v1.csv', header='column_names')
    else: # else it exists so append without writing the header
        dataforanalysis.to_csv('/content/drive/My Drive/FromColab/' + 'AA___Summary_TS2v1.csv', mode='a', header=False)

    MyMax= scaledbackprediction[scaledbackprediction.columns[0]].max()
    MyMin= scaledbackprediction[scaledbackprediction.columns[0]].min()

    Mylastactual = (Originaldata.iloc[NumberofHistoricalrecords-1,1])  #Last closing!
    print("Last actuals:")
    print(Mylastactual)


    if Mylastactual < MyMin:
        MyMin = Mylastactual

    if Mylastactual > MyMax:
        MyMax = Mylastactual

    # round the value to 2 digits after comma
    MyMin= int(100 * MyMin) / 100
    MyMax= int(100 * MyMax) / 100
   
    if MyMin == 0:
        Myvolatility = 999                
    else:
        Myvolatility = int(100 * (MyMax - MyMin) / MyMin) / 100

    #print(Myvolatility)
    
    Myevolution = (scaledbackprediction.iloc[n_steps_out - 1,0] - Originaldata.iloc[NumberofHistoricalrecords-1,0]) / Originaldata.iloc[NumberofHistoricalrecords-1,0] * 10000
    Mytrend = int(Myevolution)
    
    Myoverview = pd.DataFrame([[Mynameofcolumntopredict, Mytime, Mylastactual, MyMin, MyMax, myacc2, Mytrend]],columns=['MyIndex','Timestamp','Last actual', 'MyMin', 'MyMax', 'Accuracy','Trend'])    

    if not os.path.exists('/content/drive/My Drive/FromColab/' + "AA___Overview_TS2v1.csv"):
        Myoverview.to_csv("/content/drive/My Drive/FromColab/" + 'AA___Overview_TS2v1.csv', header='column_names')
    else: # else it exists so append without writing the header
        Myoverview.to_csv("/content/drive/My Drive/FromColab/" + 'AA___Overview_TS2v1.csv', mode='a', header=False)
    #=== make a summary file ===   

    # p=======  print and save predictions =============
    # print(scaledbackprediction)
    now=dt.datetime.now()
    mytimestamp = now.strftime("%Y%m%d%H%M") 
    myfilename = mytimestamp + "_" + MyinvestigatexIndex     
    myfilename = myfilename.replace(".","")
    myfilename = myfilename.replace("^","")
    myfilename = myfilename + "_prediction_TS2v1"+ ".csv"

    # np.savetxt("/content/drive/My Drive/FromColab/" + myfilename, scaledbackprediction, delimiter=",")

    # shutil.copy2(myfilename, '/content/drive/My Drive/FromColab/' + myfilename) 

    # ===============   Visualising the results
    #print("=== X-axis to prediction ===")
    My_Xpredictions= np.arange(NumberofHistoricalrecords,NumberofHistoricalrecords + n_steps_out ,1)
    My_Xpredictions= np.transpose(My_Xpredictions)
    
    # ======= graph #1 # Visualising the results (normalised)
    fig, ax = plt.subplots(1,2)
    plt.style.use('classic')
    
    # ======= graph #1 # Visualising the results (scaled back with Actuals)
    MyActuals = Originaldata
    MyActuals = (MyActuals.iloc[:, 1:2].values)  # Plot the first  variable
    
    ax[0].plot(MyActuals, color = 'blue', label = 'Actuals')
    ax[0].plot(My_Xpredictions, scaledbackprediction, color = 'red', label = 'Predicted Values')
    ax[0].set_title('Predictions (scaled back)')
    ax[0].set_xlabel('Time')
    ax[0].set_ylabel(Mynameofcolumntopredict)
    ax[0].set_xlim(NumberofHistoricalrecords -  10 * n_steps_out, NumberofHistoricalrecords + 2 * n_steps_out)
    # ax[0].legend()
    

    # ======= graph #2 # Visualising the results w/o Actuals
    ax[1].plot(MyActuals, color = 'blue', label = 'Actuals')
    ax[1].plot(My_Xpredictions, scaledbackprediction, color = 'red', label = 'Predicted Values')
    ax[1].set_title('Predictions (scaled back)')
    ax[1].set_xlabel('Time')
    ax[1].set_ylabel(Mynameofcolumntopredict)
    ax[1].set_xlim(NumberofHistoricalrecords -  2 * n_steps_out, NumberofHistoricalrecords + 1 * n_steps_out)
    # ax[1].legend()
    # =========save the graph ================================

    now=dt.datetime.now()
    graphname = now.strftime("%Y%m%d%H%M") 
    mygraphname = Mynameofcolumntopredict.replace(".","")
    mygraphname = mygraphname.replace("^","")
    graphname =  graphname + "_TS2v1_" + mygraphname  
    graphname =  "/content/drive/My Drive/FromColab/" + graphname
    
    plt.savefig(graphname, bbox_inches='tight')  # plt.savefig(graphname, bbox_inches='tight')
    # =======================================================
    plt.show()
    #plt.close('all')  # close all figures
    # =======================================================
    return
    #==============================================================

#=================================================================================================================================================================
#== End of big sub-routine  ======================================================================================================================================
#=================================================================================================================================================================









#== Main programm ============================================================
#my_list = ['DXS3.DE','GT','IVV','SAP.DE','APD','MUV2.DE','SCCO','EPAM','CL','GOOG','AMZN','VRSN','IP','T','MSFT','KO','TXN','CDNS','PG','V','MMM','WM','IT','VRTX','MO','BGSF','CON.DE','DB']
my_list = ['DXS3.DE','GT','SAP.DE','MUV2.DE','CON.DE']
NumberofNeurons = 200    # 200
n_steps_in = 10         # = timesteps = sequence length
n_steps_out = 10        # = how many periods do you want to predict
TrainingSets = 150       # 250       #= Epochs = how many sets do we train? The more the better. 100 is good. 
Percentageofdatasetfortraining = 80
mybatchsize = 16



# ========= delete old data ===============
for currentFile in glob.glob('/content/*.*'):
    extensions = ('.png', '.csv')
    if any(currentFile.endswith(ext) for ext in extensions):
        os.remove(currentFile)
        print("Old files deleted")
# ========= delete old data ===============

# ========= Run the main module ===============
my_list_len = len(my_list)
print(my_list_len)
for i in range(0, my_list_len):
    Predictmyshare(my_list[i])
    print(' ================================================================= ')
    print(' =================================================================')
    print(">>>>             Processing completion reate % : ")
    print(int(100 * (i+1) / my_list_len   ))
    print(' ================================================================= ')
    print(' =================================================================')
    # remount the drive jsut to be sure it is still active
    from google.colab import drive
    drive.mount('/content/drive', force_remount=True)
    # remount the drive jsut to be sure it is still active // END //
# ========= Run the main module ===============

print("=== Job done!! END===")










```
